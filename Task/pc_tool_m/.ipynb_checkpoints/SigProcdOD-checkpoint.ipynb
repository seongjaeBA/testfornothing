```python
def SigProcdOD(self, fi, MAR_flag):

global Fs ch_num
global dataSet

h = waitbar(0,'Please wait...','Name','Optical Density Calculation','Color','White',...
            'CreateCancelBtn',...
            'setappdata(gcbf,''canceling'',1)');
setappdata(h,'canceling',0)
warning('off','MATLAB:HandleGraphics:ObsoletedProperty:JavaFrame');
javaFrame = get(h,'JavaFrame');
javaFrame.setFigureIcon(javax.swing.ImageIcon('icon.png'));
wbc = allchild(h);
wbc(1).JavaPeer.setForeground( wbc(1).JavaPeer.getBackground.darkGray)
wbc(1).JavaPeer.setStringPainted(true)
SigProc = dataSet{fi}.SigProc;
RejectCh = dataSet{fi}.RejectCh;
n_data = dataSet{fi}.n_data;
waitbar(1 / 5)
    
if (SigProc.MAR == 1 && MAR_flag == 1)
   demod_data_780 = zeros(ch_num,n_data); %KJM 18.08.09
   demod_data_850 = zeros(ch_num,n_data); %KJM 18.08.09
   for i = 1 : ch_num
       if RejectCh(i) == 0
           demod_data_780(i,:) = dataSet{fi}.TimeSeries.rawData.D780(i,:);
           demod_data_850(i,:) = dataSet{fi}.TimeSeries.rawData.D850(i,:);
       elseif RejectCh(i) == 1
           demod_data_780(i,:) = 0.1;
           demod_data_850(i,:) = 0.1;
       end
   end
else
    if SigProc.SR == 0
        demod_data_780 = zeros(ch_num,n_data);
        demod_data_850 = zeros(ch_num,n_data);
        for i = 1 : ch_num
            if RejectCh(i) == 0
                demod_data_780(i,:) = dataSet{fi}.TimeSeries.fData.D780(i,:);
                demod_data_850(i,:) = dataSet{fi}.TimeSeries.fData.D850(i,:);
            elseif RejectCh(i) == 1
                demod_data_780(i,:) = 0.1;
                demod_data_850(i,:) = 0.1;
            end
        end
    elseif SigProc.SR == 1
%         time_offset = 10;
%         start_i = fix(Fs*time_offset);
%         demod_data_780 = zeros(ch_num,n_data-start_i+1);
%         demod_data_850 = zeros(ch_num,n_data-start_i+1);
        demod_data_780 = zeros(ch_num,n_data);
        demod_data_850 = zeros(ch_num,n_data);
        for i = 1 : ch_num
           if RejectCh(i) == 0
               demod_data_780(i,:) = dataSet{fi}.TimeSeries.srData.D780(i,:);
               demod_data_850(i,:) = dataSet{fi}.TimeSeries.srData.D850(i,:);
           elseif RejectCh(i) == 1
               demod_data_780(i,:) = 0.1;
               demod_data_850(i,:) = 0.1;
           end
        end
    end
end
waitbar(2 / 5)

for i = 1 : ch_num
    if (min(demod_data_780(i,:)) <= 0 || min(demod_data_850(i,:)) <= 0)
        demod_data_780(i,:) = 0.1;
        demod_data_850(i,:) = 0.1;
        RejectCh(i) = 1;
    end
end
        

% if SigProc.SR ~= 1
    if SigProc.dODBase == 0
        baselines = SigProc.sBase;
        for i = 1 : 204
            demod_data_780_ini(i) = mean(demod_data_780(i, round(baselines * Fs))); 
            demod_data_850_ini(i) = mean(demod_data_850(i, round(baselines * Fs))); 
        end
    elseif SigProc.dODBase == 1
        baselines = SigProc.sBase;
        baselinef = SigProc.eBase;
        for i = 1 : 204
            demod_data_780_ini(i) = mean(demod_data_780(i, round(baselines * Fs)+1 : round(baselinef * Fs))); 
            demod_data_850_ini(i) = mean(demod_data_850(i, round(baselines * Fs)+1 : round(baselinef * Fs))); 
        end
    end
    waitbar(3 / 5)

    for i = 1 : ch_num
        dOD_780(i,:) = -log(demod_data_780(i,:)/demod_data_780_ini(i));
        dOD_850(i,:) = -log(demod_data_850(i,:)/demod_data_850_ini(i));
    end
% else
%     for i = 1 : ch_num
%         dOD_780(i,:) = -log(demod_data_780(i,:))/log(10);
%         dOD_850(i,:) = -log(demod_data_850(i,:))/log(10);
%     end
% end
waitbar(5 / 5)

dOD_780 = real(dOD_780);
dOD_850 = real(dOD_850);
    

% dOD_780(:, 1) = 0;
% dOD_850(:, 1) = 0;



dataSet{fi}.TimeSeries.dOD.D780 = dOD_780;
dataSet{fi}.TimeSeries.dOD.D850 = dOD_850;
delete(h);
dataSet{fi}.ProcInfo.dOD = 1;
```
