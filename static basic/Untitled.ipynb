```python
from scipy import stats
import numpy as np
```

```python
def ttest_single (treatment,control,a=0.05):
  """ define and interperate a two independent t-test """
  
  # calculate degrees of freedom
  df = (len(treatment)) - 1
  # calculate critical value
  cv = stats.t.ppf((1.0 - a), df)
  # calculate t-statistic and p-value
  t_stat,p_value=stats.ttest_1samp(treatment,control)

  print('[t_stat=%.3f, p_value=%.4f, a=%.2f, df=%.3f, cv=%.3f]'%(t_stat,p_value,a,df,cv))

  if abs(t_stat) <= cv:
    print('Critical value interperation: Accept null hypothesis that the means are equal.') 
  else: 
    print('Critical value interperation: Reject the null hypothesis that the means are equal.')

  if p_value > a:
    print('P-value interepetaion: Accept null hypothesis that the means are equal.')
  else: 
    print('P-value interepetaion: Reject the null hypothesis that the means are equal.')

### ONLY CHANGE THIS PART ###

# define the test group (change based on your data) and the population mean for the 
treatment=[128, 127, 118, 115, 144, 142, 133, 140, 132, 131, 111, 132, 149, 122, 139, 119, 136, 129, 126, 128]
control=[120] 

# run t-test
ttest_single(treatment,control) # Add 'a' only if significance level is different than 0.05 (confidence level other than 95%)
```

```python
# Paired sample t-test

### TTEST FUNCTION & INTERPRETATION: DO NOT CHANGE ###

def ttest_paired (treatment,control,a=0.05):
  """ define and interprete a two independent t-test """
  
  # calculate degrees of freedom
  df = (len(treatment)) - 1            # Similar to single sample test since we're looking at two related samples. For example: before and after test.
  # calculate critical value
  cv = stats.t.ppf((1.0 - a), df)
  # calculate t-statistic and p-value
  t_stat,p_value=stats.ttest_rel(control,treatment)

  print('[t_stat=%.3f, p_value=%.4f, a=%.2f, df=%.3f, cv=%.3f]'%(t_stat,p_value,a,df,cv))

  if abs(t_stat) <= cv:
    print('Critical value interpretation: Accept null hypothesis that the means are equal.') 
  else: 
    print('Critical value interpretation: Reject the null hypothesis that the means are equal.')

  if p_value > a:
    print('P-value interpretation: Accept null hypothesis that the means are equal.')
  else: 
    print('P-value interpretation: Reject the null hypothesis that the means are equal.')

### ONLY CHANGE THIS PART ###

# define the test group (change based on your data) and the population mean for the 
treatment=[9.9, 7.9, 7.6, 6.8, 7.1, 9.9, 10.5, 9.7, 10.9, 8.2]
control=[8.0, 7.1, 6.5, 6.7, 7.2, 5.4, 4.7, 8.1, 6.3, 4.8] 

# run t-test
ttest_paired(treatment,control) # Add 'a' only if significance level is different than 0.05 (confidence level other than 95%)
```

```python
### TTEST FUNCTION & INTERPRETATION: DO NOT CHANGE ###

def ind_ttest (treatment,control,a=0.05):
  """ define and interprete a two independent t-test """
  
  # calculate degrees of freedom
  df = (len(control) + len(treatment)) - 2
  # calculate critical value
  cv = stats.t.ppf((1.0 - a), df)
  # calculate t-statistic and p-value
  t_stat,p_value=stats.ttest_ind(treatment,control)

  print('[t_stat=%.3f, p_value=%.3f, a=%.2f, df=%.3f, cv=%.3f]'%(t_stat,p_value,a,df,cv))

  if abs(t_stat) <= cv:
    print('Critical value interpretation: Accept null hypothesis that the means are equal.') 
  else: 
    print('Critical value interpretation: Reject the null hypothesis that the means are equal.')

  if p_value > a:
    print('P-value interpretation: Accept null hypothesis that the means are equal.')
  else: 
    print('P-value interpretation: Reject the null hypothesis that the means are equal.')

### ONLY CHANGE THIS PART ###

# define the test groups (change based on your data)
treatment=[118, 115, 112, 120, 124, 130, 123, 110, 120, 121,123, 125, 129, 130, 112, 117, 119, 120, 123, 128]
control=[128, 127, 118, 115, 144, 142, 133, 140, 132, 131,111, 132, 149, 122, 139, 119, 136, 129, 126, 128] 

# run t-test
ind_ttest(treatment,control) # Add 'a' only if significance level is different than 0.05 (confidence level other than 95%)
```
