```python
def age_to_numeric(x):
    if x=='초1':
        return 1
    if x=='초2':
        return 2
    if x=='초3':
        return 3
    if x=='초4':
        return 4
    if x=='초5':
        return 5
    if x=='초6':
        return 6
    if x=='중1':
        return 7
    
def ID_to_IDNum(x):
    temp = str(x).strip().split('_')
    if len(temp) < 2:
        return np.nan
    return int(temp[2])
```

```python
list_child = pd.read_csv('./data/list.csv', encoding = "euc_kr", na_values = 'empty')
list_child['IDNum'] = list_child['UID'].apply(ID_to_IDNum)
list_child['AgeNum'] = list_child['Age'].apply(age_to_numeric)

data_fnirs = pd.read_csv('./data/fnirs_feature_0322.csv', encoding = "euc_kr", na_values = 'empty')
data_fnirs['age_num'] = data_fnirs['fnirs_age'].apply(age_to_numeric)
data_fnirs.rename({'fnirs_idnum' : 'IDNum'}, inplace = True, axis = 'columns')

# data_fnirs = pd.read_csv('./data/ex_csv.csv', encoding = "euc_kr", na_values = 'empty')
# data_fnirs['age_num'] = data_fnirs['fnirs_basic_age'].apply(age_to_numeric)
# data_fnirs.rename({'fnirs_idnum' : 'IDNum'}, inplace = True, axis = 'columns')

data_meta = pd.read_excel('./data/total_meta.xlsx', encoding = "euc_kr", na_values = 'empty')
data_meta['IDNum'] = data_meta['아동연구번호'].apply(ID_to_IDNum)
data_meta['AgeNum'] = data_meta['학년_1'].apply(age_to_numeric)
```
